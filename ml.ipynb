```python
import pandas as pd
cc_apps = pd.read_csv('cc_approvals.data', header = None)
cc_apps.head()
```

```python

cc_apps_description = cc_apps.describe()
print(cc_apps_description)
print("\n")
cc_apps_info = cc_apps.info()
print(cc_apps_info)
print("\n")
cc_apps.tail(17)
```

```python

import numpy as np
print(cc_apps.tail(17))
cc_apps = cc_apps.replace('?', np.nan)
print(cc_apps.tail(17))
```

```python

cc_apps.fillna(cc_apps.mean(), inplace=True)
cc_apps.isnull().sum()
```

```python

for col in cc_apps.columns:
    if cc_apps[col].dtypes == 'object':
        cc_apps[col] = cc_apps[col].fillna(cc_apps[col].value_counts().index[0])
print(cc_apps.isnull().sum().sum())
```

```python

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in cc_apps.columns:
    if cc_apps[col].dtype=='object':
        cc_apps[col]=le.fit_transform(cc_apps[col].values)
```

```python

from sklearn.model_selection import train_test_split

cc_apps = cc_apps.drop([11, 13], axis=1)
cc_apps = cc_apps.values
X,y = cc_apps[:,0:-2] , cc_apps[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.33,
                                                    random_state=42)
```

```python

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX_train = scaler.fit_transform(X_train)
rescaledX_test = scaler.fit_transform(X_test)
```

```python

from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(rescaledX_train, y_train)
```

```python

from sklearn.metrics import confusion_matrix
y_pred = logreg.predict(rescaledX_test)
print("Accuracy of logistic regression classifier: ", logreg.score(rescaledX_test, y_test))
confusion_matrix(y_test, y_pred)
```

```python

from sklearn.model_selection import GridSearchCV
tol = [0.01, 0.001, 0.0001]
max_iter = [100, 150, 200]
param_grid = dict({'tol': tol, 'max_iter': max_iter})
```

```python

grid_model = GridSearchCV(estimator=logreg, param_grid=param_grid, cv=5)
rescaledX = scaler.fit_transform(X)
grid_model_result = grid_model.fit(rescaledX, y)
best_score, best_params = grid_model_result.best_score_, grid_model_result.best_params_
print("Best: %f using %s" % (best_score, best_params))
```

```python
import numpy as np
new_data = {
    'column1': 'b',
    'column2': 10.0,
    'column3': 0.5,
    'column4': 'u',
    'column5': 'g',
    'column6': 'c',
    'column7': 'v',
    'column8': 0.2,
    'column9': 'f',
    'column10': 'f',
    'column12': 0,
    'column14': 0.0,
}
column_order = ['column1', 'column2', 'column3', 'column4', 'column5', 'column6', 'column7', 'column8', 'column9', 'column10', 'column12', 'column14']

label_encoders = {}
data = np.zeros((1, len(column_order)))

for col_index, col in enumerate(column_order):
    if col in label_encoders:

        data[0, col_index] = label_encoders[col].transform([new_data[col]])[0]
    else:
        if isinstance(new_data[col], str):

            label_encoders[col] = LabelEncoder()
            data[0, col_index] = label_encoders[col].fit_transform([new_data[col]])[0]
        else:

            data[0, col_index] = new_data[col]

data = scaler.transform(data)

prediction = logreg.predict(data)
print("Predicted class:", prediction[0])
```
